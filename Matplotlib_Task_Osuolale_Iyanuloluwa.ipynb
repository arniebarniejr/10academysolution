## Question 1
#### Making a Box Plot 

```python
Box1 = df[df['Q2']== 1]["Longitude"]
Box2 = df[df['Q2']== 2]['Longitude']
plt.style.use("ggplot")
fig, ax = plt.subplots()
ax.boxplot((Box1, Box2),vert=True, showmeans=True, meanline=True,
           labels=("Male", "Female"), patch_artist=True,
           medianprops= {'linewidth': 2, "color": "green"}, meanprops= {'linewidth' : 2, "color": "red"})
plt.xlabel("Q2(GENDER)")
plt.ylabel('Longitude')
plt.title('A well prepared Boxplot for Q2(GENDER) against Longitude')
plt.show()
```

## Question 2
Making a Bar Chart 

```python

age = df['Q2'].value_counts().tolist()
f, ax = plt.subplots(figsize=(7,5)) 
gender = ['Female', 'Males']
plt.title("A Bar plot showing the number of Males and Females")
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.bar(gender, age)
plt.show()
```

## Question 3
#### A scatter plot

```python
plt.scatter(df["Latitude"], df['Longitude'], color='red', edgecolors='black')
plt.title("A scatter plot showing the Latitude and Longitude", size=15)
plt.xlabel("Latitude", size=15)
plt.ylabel("Longitude", size=15);
plt.show()
```

## Q4 
#### An Histogram

```python
plt.hist(df['Q1'], histtype= "stepfilled", bins=10, label="stepfilled", color='green')
plt.title("A Histogram showing the Frequency distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show();
```
